```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = [('2000-2006', 'Action;Drama', 1240), 
        ('2000-2006','Adventure;Sci-Fi',67),   
        ('2000-2006', 'Comedy;Romance',1620),    
        ('2007-2013', 'Action;Drama', 1811), 
        ('2007-2013','Adventure;Sci-Fi',153),  
        ('2007-2013', 'Comedy;Romance',2402),   
        ('2014-2020','Action;Drama',2326),     
        ('2014-2020','Adventure;Sci-Fi',247),     
        ('2014-2020','Comedy;Romance',2784)]
```

```python
print (data)
```

```python
data1 = []
data2 = []
data3 = []
for d in data:
    if d[0] == '2000-2006':
        data1.append(d)
    elif d[0] == '2007-2013':
        data2.append(d)
    elif d[0] == '2014-2020':
        data3.append(d)
fig, (d1, d2, d3) = plt.subplots(1, 3, figsize=(15, 9))
d1.bar([d[1] for d in data1], [d[2] for d in data1])
d1.set_title('2000-2006')
d1.set_xticklabels([d[1] for d in data1], rotation=90)
d2.bar([d[1] for d in data2], [d[2] for d in data2])
d2.set_title('2007-2013')
d2.set_xticklabels([d[1] for d in data2], rotation=90)
d3.bar([d[1] for d in data3], [d[2] for d in data3])
d3.set_title('2014-2020')
d3.set_xticklabels([d[1] for d in data3], rotation=90)
plt.show()
```

```python
df = pd.DataFrame(data, columns=['Year', 'Genre', 'Number'])

sns.lineplot(x='Year', y='Number', data=df[df['Genre'] == 'Action;Drama'], label='Action;Drama')
sns.lineplot(x='Year', y='Number', data=df[df['Genre'] == 'Comedy;Romance'], label='Comedy;Romance')
sns.lineplot(x='Year', y='Number', data=df[df['Genre'] == 'Adventure;Sci-Fi'], label='Adventure;Sci-Fi')
plt.title('Number of Releases by Genre and Year')

plt.show()
```

```python
data1 = {}
data2 = {}
data3 = {}

for d in data:
    if d[0] == '2000-2006':
        if d[1] in data1:
            data1[d[1]] += d[2]
        else:
            data1[d[1]] = d[2]
    elif d[0] == '2007-2013':
        if d[1] in data2:
            data2[d[1]] += d[2]
        else:
            data2[d[1]] = d[2]
    elif d[0] == '2014-2020':
        if d[1] in data3:
            data3[d[1]] += d[2]
        else:
            data3[d[1]] = d[2]


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.pie(data1.values(), labels=data1.keys())
plt.title('2000-2006')

plt.subplot(1, 3, 2)
plt.pie(data2.values(), labels=data2.keys())
plt.title('2007-2013')

plt.subplot(1, 3, 3)
plt.pie(data3.values(), labels=data3.keys())
plt.title('2014-2020')


plt.show()
```
